```python
df['crash_date'] = pd.to_datetime(df['crash_date'],format='mixed',dayfirst=True,errors='coerce')
```

```python
df['Year'] = df['crash_date'].dt.year
df['Day'] = df['crash_date'].dt.day
```

```python
df['weather_condition'].fillna('UNKNOWN', inplace=True)
df['lighting_condition'].fillna('UNKNOWN', inplace=True)
df['traffic_control_device'].fillna('UNKNOWN', inplace=True)
```

```python
df = df.drop_duplicates()
```
